```python
# Import dependencies
import pandas as pd
```

```python
# Read in batting csvs
bat_22 = pd.read_csv("Resources/2022_Batting.csv")
bat_21 = pd.read_csv("Resources/2021_Batting.csv")
bat_20 = pd.read_csv("Resources/2020_Batting.csv")
bat_19 = pd.read_csv("Resources/2019_Batting.csv")
bat_18 = pd.read_csv("Resources/2018_Batting.csv")
bat_17 = pd.read_csv("Resources/2017_Batting.csv")
bat_16 = pd.read_csv("Resources/2016_Batting.csv")
bat_15 = pd.read_csv("Resources/2015_Batting.csv")
bats = [bat_15,bat_16,bat_17,bat_18,bat_19,bat_20,bat_21,bat_22]
```

```python
bat_18.head(30)
```

```python
print(bat_18.columns)
```

```python
# Drop unwanted batting columns
for bat in bats:
    bat.drop(columns=['#Bat','G','PA','BA','OBP','SLG', 'OPS', 'OPS+','TB', 'GDP','IBB'], inplace=True)
bat_18.head(2)
```

```python
# Read in pitching csvs
pit_22 = pd.read_csv("Resources/2022_Pitching.csv")
pit_21 = pd.read_csv("Resources/2021_Pitching.csv")
pit_20 = pd.read_csv("Resources/2020_Pitching.csv")
pit_19 = pd.read_csv("Resources/2019_Pitching.csv")
pit_18 = pd.read_csv("Resources/2018_Pitching.csv")
pit_17 = pd.read_csv("Resources/2017_Pitching.csv")
pit_16 = pd.read_csv("Resources/2016_Pitching.csv")
pit_15 = pd.read_csv("Resources/2015_Pitching.csv")
pits=[pit_15,pit_16,pit_17,pit_18,pit_19,pit_20,pit_21,pit_22]
pit_18.head(30)
```

```python
pit_18.columns
```

```python
# Drop unwanted pitching columns
for pit in pits:
    pit.drop(columns=['#P', 'ERA', 'G', 'GS', 'GF','CG', 'cSho', 'IP', 'ER', 'BK', 'WP','ERA+', 'FIP', 'WHIP', 'H9', 'HR9','BB9', 'SO9', 'SO/W'], inplace=True)
pit_18.head(2)
```

```python
# Read in fielding csvs
def_22 = pd.read_csv("Resources/2022_Fielding.csv")
def_21 = pd.read_csv("Resources/2021_Fielding.csv")
def_20 = pd.read_csv("Resources/2020_Fielding.csv")
def_19 = pd.read_csv("Resources/2019_Fielding.csv")
def_18 = pd.read_csv("Resources/2018_Fielding.csv")
def_17 = pd.read_csv("Resources/2017_Fielding.csv")
def_16 = pd.read_csv("Resources/2016_Fielding.csv")
def_15 = pd.read_csv("Resources/2015_Fielding.csv")
fields = [def_15,def_16,def_17,def_18,def_19,def_20,def_21,def_22]
def_18.head(2)
```

```python
def_18.columns
```

```python
# Drop unwanted fielding columns
for field in fields:
    field.drop(columns=['#Fld', 'RA/G', 'G', 'GS', 'CG', 'Inn', 'Fld%', 'Rtot', 'Rtot/yr', 'Rdrs', 'Rdrs/yr', 'Rgood'], inplace=True)
def_18.head(2)
```

```python
# Merge dataframes by year
merged_22 = pd.merge(bat_22, pit_22, on='Tm')
merged_22 = pd.merge(merged_22, def_22, on = 'Tm')

merged_21 = pd.merge(bat_21, pit_21, on='Tm')
merged_21 = pd.merge(merged_21, def_21, on = 'Tm')

merged_20 = pd.merge(bat_20, pit_20, on='Tm')
merged_20 = pd.merge(merged_20, def_20, on = 'Tm')

merged_19 = pd.merge(bat_19, pit_19, on='Tm')
merged_19 = pd.merge(merged_19, def_19, on = 'Tm')

merged_18 = pd.merge(bat_18, pit_18, on='Tm')
merged_18 = pd.merge(merged_18, def_18, on = 'Tm')

merged_17 = pd.merge(bat_17, pit_17, on='Tm')
merged_17 = pd.merge(merged_17, def_17, on = 'Tm')

merged_16 = pd.merge(bat_16, pit_16, on='Tm')
merged_16 = pd.merge(merged_16, def_16, on = 'Tm')

merged_15 = pd.merge(bat_15, pit_15, on='Tm')
merged_15 = pd.merge(merged_15, def_15, on = 'Tm')
```

```python
# Rename team names to include year
merged_22.Tm = merged_22.Tm +'_2022'
merged_21.Tm = merged_21.Tm +'_2021'
merged_20.Tm = merged_20.Tm +'_2020'
merged_19.Tm = merged_19.Tm +'_2019'
merged_18.Tm = merged_18.Tm +'_2018'
merged_17.Tm = merged_17.Tm +'_2017'
merged_16.Tm = merged_16.Tm +'_2016'
merged_15.Tm = merged_15.Tm +'_2015'
```

```python
# Merging yearly dataframes into single dataframe
mlb_complete = pd.concat([merged_22,merged_21,merged_20,merged_19,merged_18,merged_17,merged_16,merged_15])
```
